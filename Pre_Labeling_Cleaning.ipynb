# Import Library

```python
import pandas as pd
```

# Pre-Labeling Data Cleaning

```python
# Read the Data from CSV File
df = pd.read_csv("/content/comments.csv", sep=';')

print(f"Total row:  {len(df)}")
```

```python
# Remove Null Row
df = df.dropna(subset="text")

print(f"Total Row After Removing Null Comments: {len(df)}")

```

```python
# Remove Duplicated Data
df.drop_duplicates(inplace=True)

print(f"Total Row After Removing Duplicated Comments: {len(df)}")
```

```python
# Remove comments with text lower than 1 word

df["word_count"] = df["text"].apply(lambda x: len(str(x).split()))

df_filtered = df[df["word_count"] > 1].copy()

df_filtered = df_filtered.drop(columns=["word_count"])

print(f"Total Row Ready for Annotation: {len(df_filtered)}")
```

```python
# Export to CSV

df_filtered.to_csv("comments_for_annotation.csv", index=False, sep=';')

print("File 'comments_for_annotation.csv' has been exported")
```
